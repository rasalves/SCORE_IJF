```r
library(reticulate)
library(RColorBrewer)
library(readr)

np <- import("numpy")
load("RES/INTERPRETABILITY/clusters.RData")
TSNE <- np$load("RES/INTERPRETABILITY/TSNE_INT.npy")
PRED <- np$load("RES/INTERPRETABILITY/TSNE_PRED_INT.npy")


groupEvents = list()
groupEvents[[1]]= c(1)
groupEvents[[2]]= c(2,3,4)
groupEvents[[3]]= c(5)
groupEvents[[4]]= c(6,7)
groupEvents[[5]]= c(8)
groupEvents[[6]]= c(9)
NG = c("Goal","Attempts","Corner","Cards","Substitution","Offside")

load("clusters.RData")

matchesDF <- as.matrix(read_csv("DATA/matchesDF.csv"))  
match = list()
for(i in 1:dim(matchesDF)[1]) match[[matchesDF[i,7]]] = paste(matchesDF[i,3],matchesDF[i,5],"X",matchesDF[i,6],matchesDF[i,4],sep =" ")
```

```r
SAVE_TO_FILE = FALSE

if (SAVE_TO_FILE) pdf(file = "figure8.pdf",width = 13, height = 18) 

lt = rbind(matrix(1,ncol=6,nrow=4),rbind(2:7,8:13,14:19))
nf <- layout( lt)
par(mar=c(5,7,2,0.1),oma = c(0.2,0.2,0.2,0.2))

mainAn = list()
mainAn[[3]] = c(1325,828,781,1105,2732,2735,3803,4078,7276,6020)
mainAn[[3]] =c(1105,2735,6020)
for(C in 3){
    coltimeAn = colTime
    coltimeAn[coltimeAn!=C] = 7
    plot(TSNE,col=colors_list_time[coltimeAn],cex=0.95,pch=16,xlab="",ylab="",cex.axis=2.8)
    mtext(text="t-SNE DIM 1",side=1,cex =2.5,line=4.2,outer=F)
    mtext(text="t-SNE DIM 2",side=2,cex =2.5,line=4.2,outer=F)
    
    clsmp = colTime
    X = TSNE[clsmp==C,]
    mX = (PRED[,1])[clsmp==C]
    centroid = colMeans(X)
    
    
    points(centroid[1],centroid[2],pch = 16, cex = 5,col = "navy")
    points(centroid[1],centroid[2],pch = 1, cex = 6.2,col = "navy")
    
    center <- colMeans(X)
    cov_matrix <- cov(X)
    mahalanobis_distances <- mahalanobis(X, center, cov_matrix)
    n = length(mahalanobis_distances)
    th = sort(mahalanobis_distances)[round(0.95 * n)+1]
    
    points(X[mahalanobis_distances>th,],cex=1,pch=16, col="red")
    mAn = (PRED[clsmp==C,])[mahalanobis_distances>th,1]
    colan2 = c("#76c3a0","#FFB703","#76d9e0")
    a = 1
    for(m in  mainAn[[C]]){
        x = X[mX==m,1]
        y = X[mX==m,2]
        points(x,y,cex=5,pch=16, col=colan2[a])
        points(x,y,cex=6.2,pch=1, col=colan2[a])
        a =  a + 1
    }
    
}

a = 1
for(m in mainAn[[3]]){
    meanC = colMeans(PRED[clsmp==C,-c(1,2)])
    for(i in 1:length(groupEvents)){
        RG = c()
    
        PC = PRED[clsmp==C,-c(1,2)]
        PCA = PRED[clsmp==C & PRED[,1] == m,-c(1,2)]
        
        H = mean(PC[,groupEvents[[i]]])
        A = mean(PC[,groupEvents[[i]]+9])
    
        AH = mean(PCA[groupEvents[[i]]])
        AA = mean(PCA[groupEvents[[i]]+9])
        
        R = c(H,AH,A,AA)
        RG = c(RG,R)
    
        colan = c(colors_list_time[unique(coltimeAn[coltimeAn!=7])],colan2[a],colors_list_time[unique(coltimeAn[coltimeAn!=7])],colan2[a])
        barplot(RG * 100,col = colan ,border = colan,density=rep(c(100,100,30,30),8),name = "",cex.axis=2.5)
        box()
        mtext(text=NG[i],side=1,cex =2.0,line=2,outer=F)
        if(i==1 && a == 2){
            if(i == 1) mtext(text="Probability (%)",side=2,cex =2.5,line=4.2,outer=F)
        }
        
    }
    print(paste(m,match[[m]],sep=" "))#,side=3,cex =1,line=-1,outer=T)
    a = a + 1
}

if (SAVE_TO_FILE) dev.off()
```

```r
SAVE_TO_FILE = FALSE

if (SAVE_TO_FILE) pdf(file = "figure9.pdf",width = 18, height = 30) 

pointsMatch<-function(m,colline,P,lty=1){
    points(TSNE[P[,1] == m,],lwd=5,type = 'l',col=colline,lty=lty)
    points(TSNE[P[,1] == m,],cex = 5,pch = 16,col = colline)
    points((TSNE[P[,1] == m,])[1,1],(TSNE[P[,1] == m,])[1,2],cex = 2,pch = 16,col = "white")
    points((TSNE[P[,1] == m,])[1,1],(TSNE[P[,1] == m,])[1,2],cex = 2,pch = 1,col = "black")
    points((TSNE[P[,1] == m,])[6,1],(TSNE[P[,1] == m,])[6,2],cex = 2,pch = 16,col = "black")
    return (paste(m, match[[m]],colline))
}

pointsMatchDashed<-function(m,colline,P){
    points(TSNE[P[,1] == m,][1:3,],lwd=5,type = 'l',col=colline)
    points(TSNE[P[,1] == m,][1:3,],cex = 5,pch = 16,col = colline)
    if(colline=="reds"){
        points(TSNE[P[,1] == m,][3:6,],lwd=5,type = 'l',col=colline)
        points(TSNE[P[,1] == m,][3:6,],cex = 5,pch = 16,col = colline)
    }else{
        points(TSNE[P[,1] == m,][3:6,],lwd=5,lty=2,type = 'l',col= colline )
        points(TSNE[P[,1] == m,][3:6,],cex = 5,pch = 16,col = colline)
    }
    points((TSNE[P[,1] == m,])[1,1],(TSNE[P[,1] == m,])[1,2],cex = 2,pch = 16,col = "white")
    points((TSNE[P[,1] == m,])[1,1],(TSNE[P[,1] == m,])[1,2],cex = 2,pch = 1,col = "black")
    points((TSNE[P[,1] == m,])[6,1],(TSNE[P[,1] == m,])[6,2],cex = 2,pch = 16,col = "black")
    return (paste(m, match[[m]],colline))
}

nf <- layout( matrix(c(1:8), nrow=4,byrow=TRUE) )
par(mar=c(12,6,0,0.2),oma = c(0.2,0.2,0.2,0.2) )

rec = list()
rec[[625]] = c(6892,830,8054)
rec[[4385]] = c(434,9444,7707)
rec[[7010]] = c(4845,2798,5547)
rec[[8067]] = c(6918,7729,7926)
rec[[9912]] = c(3797,4066,8226)
rec[[9970]] = c(2742,4053,8158)
rec[[9993]] = c(5896,947,4070)
rec[[10004]] = c(9276,1988,398)

cols = c("#2a7e57","#FFB703","#2046b0")
msec = c(7010,4385,8067,625)
for(m1 in msec){
    a = 1
    plot(TSNE,col=colors_list[clusters],cex=0.45,pch=16,xlab="",ylab="",cex.axis=1.8)
    mtext(text="t-SNE DIM 1",side=1,cex =1.5,line=2.8,outer=F)
    mtext(text="t-SNE DIM 2",side=2,cex =1.5,line=2.8,outer=F)
    l=""
    for(m in rec[[m1]]){ 
        l = paste(pointsMatch(m,cols[a],PRED,lty=1),sep="")
        print(l)
        a = a + 1
    }
    l = paste(a,pointsMatch(m1,"#bf132a",PRED))
    print(l)
    print("")
}


msec  = c(9912,9970,9993,10004)
for(m1 in msec){
    plot(TSNE,col=colors_list[clusters],cex=0.45,pch=16,xlab="",ylab="",cex.axis=1.8)
    mtext(text="t-SNE DIM 1",side=1,cex =1.5,line=2.8,outer=F)
    mtext(text="t-SNE DIM 2",side=2,cex =1.5,line=2.8,outer=F)
    a = 1
    for(m in rec[[m1]]){
        l = paste(pointsMatchDashed(m,cols[a],PRED)," ",m1," ",rec[[m1]][a],sep="")
        print(l)
        a = a + 1
    }
    l1 = pointsMatchDashed(m1,"#bf132a",PRED)
    print(l1)
    print("")
}

if (SAVE_TO_FILE) dev.off()
```
