```r
SAVE_TO_FILE = FALSE

library(reticulate)
library(readr)
np <- import("numpy")
x <- np$load("RES/FIGURES/RR.npy")

color_list <- c("#e3342f", "#f6993f", "#ffed4a", "#38c172", "#4dc0b5", "#3490dc", "#6574cd", "#9561e2", "#f66d9b")

if (SAVE_TO_FILE) pdf(file = paste("figure2.pdf",sep=""),width = 12, height = 6) 

#Margin and layout
nf <- layout(matrix(1:2,nrow=1,byrow=T) )
par(mar=c(0.2,0.2,2,0.2),oma = c(4,4,0.2,0.2) )

#plot an labels home team
plot(c(),xlim=c(1,20),ylim=c(0,1),xaxt='n',cex.axis = 1.3,xlab="",ylab="",main="Home Team")
ax = 1:18
mtext(text=paste("Rank",sep=""),side=1,cex =1.6,line=2.5,outer=T)
mtext(text=paste("Relative Cumulative Frequency",sep=""),side=2,cex =1.6,line=2.5,outer=T)
par(las=2)
axis(side = 1, at = ax, labels = ax,cex.axis=1.3)

a = 1
for(j in 0:8){
    y = c()
    for(i in 1:18){
        f = sum((1/x[x[,2]==j,1])<i)
        y = c(y,f)
    }
    
    points(1:18,y/max(y),type='l',main = j,col = color_list[a],lwd = 2)
    a = a + 1
}


#plot an labels away team
plot(c(),xlim=c(1,20),ylim=c(0,1),xaxt='n',xlab="",ylab="",yaxt='n',main="Away Team")
ax = 1:18
par(las=2)
axis(side = 1, at = ax, labels = ax,cex.axis=1.3)
a = 1
for(j in 9:17){
    y = c()
    for(i in 1:18){
        f = sum((1/x[x[,2]==j,1])<i)
        y = c(y,f)
    }
    
    points(1:18,y/max(y),type='l',main = j,col = color_list[a],lwd = 2)
    a = a + 1
}

#legend
ne = c("GOAL","ATON","AOFF","ABLK","COKC","YLCA","RDCA","SUBS","OFFS")
legend("bottomright", legend = ne, fill = color_list, title = "", bg = "transparent",bty = "n")
if(SAVE_TO_FILE) dev.off()
```
