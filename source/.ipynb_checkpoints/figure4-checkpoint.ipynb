```r
SAVE_TO_FILE = FALSE


if (SAVE_TO_FILE) pdf(file = paste("FIGURES/figure4.pdf",sep=""),width = 12 , height = 20)  

options(warn = -1)
library(reticulate)
library(readr)

#Plot Functions
meanPloat<-function(xs,M,colBorder,linCol,lim,event="",den = TRUE){
    STD1 = M[1,xs] + M[2,xs]
    STD2 = M[1,xs] - M[2,xs]
    STD1[STD1<0]=0
    STD2[STD2<0]=0
    if(max(xs) == 90){
        STD1[length(STD1)] = STD1[length(STD1)-1]
        STD2[length(STD2)] = STD2[length(STD2)-1]
    }
    if(den){
        points(xs,M[1,xs], ylim = c(0,lim),main = event,type ="l",col = linCol,lwd = 2)
    }else{
        points(xs,M[1,xs], ylim = c(0,lim),main = event,type ="l",col = linCol,lwd = 0.1)
        points(xs,M[1,xs], ylim = c(0,lim),main = event,type ="l",col = linCol,lwd = 3, lty = 2)
    }
    for(j in 1:(length(STD1)-1)){
        x <- c(xs[j],xs[j+1],xs[j+1],xs[j])
        y <- c(STD2[j],STD2[j+1],STD1[j+1],STD1[j])
        if(den){
            polygon(x, y, col = colBorder, border = F)
            polygon(x, y, col = colBorder, border = F)
        }else{
            polygon(x, y, col = colBorder, border = F, density = 14, angle = 45)
            polygon(x, y, col = colBorder, border = F, density = 14, angle = 45)
            polygon(x, y, col = colBorder, border = F, density = 14, angle = 135)
            polygon(x, y, col = colBorder, border = F, density = 14, angle = 135)
        }
    }
    
}

smt<-function(v,k=5){
    x = v[1:(k-1)]
    for(i in k:length(v)){
        x = c(x,median(v[(i-k+1):i]))
    }
    return(x)
}


np <- import("numpy")
path = "RES/MATCHES/"

#Load data
eventslist <- as.matrix(read_csv("DATA/eventsPostProcessed.csv"))  
matchesDF <- as.matrix(read_csv("DATA/matchesDF.csv"))  
match = list()
for(i in 1:dim(matchesDF)[1]) match[[matchesDF[i,7]]] = paste(matchesDF[i,3],matchesDF[i,5],"X",matchesDF[i,6],matchesDF[i,4],sep =" ")

#Initial Parameters
results = matrix(0,ncol = 90,nrow = 18)
counts  = matrix(0,ncol = 90,nrow = 18)

events = c('GOAL','AONT','Shot off target','Shot Blocked', 'COKC', 'YLCA','Red card','SUBS','Offside')
lims = c(0.05,0.15,0.20,0.15,0.20,0.08,0.005, 0.30,0.10)


#Collecting collective behaviour
R = list()
for(i in 1:18){
    R[[i]] = rep(0,90)
}

for(i in 1:18){
    for (m  in 9659:10113){
        file = paste(path,"MATCH_",m,".npz",sep="")
        if (file.exists(file)) { 
            data <- np$load(file)
            pred = (data['pred'])[,2:length((data['pred'])[1,])]
            time = floor((data['pred'])[,1])
            tap = (tapply(pred[time<=90,i],time[time<=90], sum))
            results[i,as.numeric(names(tap))] = results[i,as.numeric(names(tap))]  + unname(tap)
            tt = table(time[time<90])
            counts[i,as.numeric(names(tt))] = counts[i,as.numeric(names(tt))] +  unname(tt)
            v = rep(0,90)
            v[as.numeric(names(tap))] = unname(tap)
            v[as.numeric(names(tt))] = v[as.numeric(names(tt))]/  unname(tt)
            R[[i]] = rbind(R[[i]],v)
        }
    }
}

RES = list()
for(i in 1:18){
    vat = (R[[i]])[-1,]
    for(j in 1:90){
        r = c(mean(vat[vat[,j]!=0,j]),sd(vat[vat[,j]!=0,j]))
        if(j == 1){
            res = r
        }else{
            res = cbind(res,r)
        }
    }
    RES[[i]] = res
}

#Margin and layout of the file
nf <- layout(matrix(1:8,nrow=4,byrow=T) )
par(mar=c(5,6,2,0.2),oma = c(0.2,0.2,0.2,0.2) )

matchesGraph = c(10109,9927,10014,9930,9765,9928,9922,10006)
z = 1
is = c(1,1,1,1,2,5,6,8)
for (m  in matchesGraph){
    file = paste(path,"MATCH_",m,".npz",sep="")
    if (file.exists(file)) {
        data <- np$load(file)
        pred = (data['pred'])[,2:length((data['pred'])[1,])]
        time = (data['pred'])[,1]
        i = is[z]
        z = z+ 1
        elim = -lims[i]/2
        xs = 1:90
        #Plotting the graphs            
        plot(c(),yaxt='n',xaxt='n',xlab="",ylab="",xlim=c(min(xs),max(xs)), ylim = c(elim,lims[i]),main.cex = 1.3,main =paste(match[[m]]),type ="l",col = "blue")
        ax = seq(0,lims[i],length.out=5)
        par(las = 0)
        mtext(text=paste("         P(",events[i],")",sep=""),side=2,cex =1.4,line=2.8,outer=F)
        ax  = floor(ax * 100)/100
        par(las = 1)
        mtext(text=paste("Time (t)",sep=""),side=1,cex =1.6,line=2.9,outer=F)
        axis(side = 1, at = c(1,15,30,45,60,75,90), labels = c(1,15,30,45,60,75,90),cex.axis=1.3)
        par(las = 3)
        axis(side = 2, at = ax, labels = ax,cex.axis=1.3)
        colBorder = colBorder = rgb(0, 0, 1, alpha = 0.05)
        meanPloat(xs,RES[[i]],colBorder,rgb(0, 0, 1, alpha = 0.25),lims[i],den=TRUE)
        colBorder  = rgb(0.2353, 0.7216, 0.2431, alpha = 0.25)
        meanPloat(xs,RES[[i+9]],colBorder,rgb(0, 1, 0, alpha = 0.50),lims[i],den=FALSE)
        abline(v = 45,lwd=2)
        xs = 5:90
        K=10
        points(time[time>=min(xs) & time<= max(xs)],smt(pred[,i],k=K)[time>=min(xs) & time<= max(xs)],type="l",col = "navy", lwd=1.5)
        points(time[time>=min(xs) & time<= max(xs)],smt(pred[,i+9],k=K)[time>=min(xs) & time<= max(xs)],type="l",col = "darkgreen", lwd=0.1)
        points(time[time>=min(xs) & time<= max(xs)],smt(pred[,i+9],k=K)[time>=min(xs) & time<= max(xs)],type="l",col = "darkgreen", lwd=3, lty = 2)
        x = data['tabEvts']
        x = x[-1,]
        
        if(sum(x[,5]==(i-1))>0){
            s = x[x[,5]==(i-1),6]
            s = s[s>=min(xs)]
            s = s[s<=max(xs)]
            points(s,(smt(pred[,i],k=K))[time %in% s],col="navy", cex=1.6,pch=16)
            points(s,(smt(pred[,i],k=K))[time %in% s],col="blue", cex=2)
        }

        if(sum(x[,5]==(i+9-1))>0){
            s = x[x[,5]==(i+9-1),6]
            s = s[s>=min(xs)]
            s = s[s<=max(xs)]
            points(s,(smt(pred[,i+9],k=K))[time %in% s],col="darkgreen", cex=1.6,pch=15)
            points(s,(smt(pred[,i+9],k=K))[time %in% s],col="green", cex=2, pch=22)
        }

        #Plotting Events
        x = data['tabEvts']
        ne = c("Goal","Shot","Foul","Corner","Card","Sub","OS")
        ne = c("GOAL","ATTEMPTS","COKC","YLCA","RDCA","SUBS","OFFS","FOUL")
        CODE = list()
        CODE[[1]] = c(0)
        CODE[[2]] = c(1,2,3,4)
        CODE[[3]] = c(5)
        CODE[[4]] = c(6)
        CODE[[5]] = c(7)
        CODE[[6]] = c(8)
        CODE[[7]] = c(9)
        CODE[[8]] = c(11)
            
        se = seq(0,elim,length.out=9)   
        par(las = 2)
        axis(side = 2, at = se[2:9], labels =ne,cex.axis=1.0)
        a =0
        e1 = eventslist[eventslist[,1]%in% m,]
        for(j in se){
            if(j == 0){
                abline(j,0)
            }else{
                
                abline(j,0,col="gray")
                jat = CODE[[a]]
                sd = 1
                e = e1[(e1[,5] %in% jat) & (e1[,4] == sd),3]
                e = e[e>=min(xs)]
                e = e[e<=max(xs)]
                if(length(e)>0){ 
                    points(e,rep(j,length(e)),col = "#03a0c9",pch=16,cex =1.3)
                }
              
                sd = 2
                e = e1[e1[,5] %in% jat & e1[,4] == sd,3]
                e = e[e>=min(xs)]
                e = e[e<=max(xs)]
                if(length(e)>0){ 
                    points(e,rep(j,length(e)),col = rgb(0.2353, 0.7216, 0.2431, alpha = 1),pch=15,cex =1.3)
                }
            }
            a = a+1
        }   
    }   
}

if (SAVE_TO_FILE) dev.off()
```
